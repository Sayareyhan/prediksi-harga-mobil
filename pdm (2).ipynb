```python
df = df.drop_duplicates()

# Membersihkan kolom Harga (hapus Rp, titik, dll)'
df['Price'] = df['Price'].astype(str)
df['Price'] = df['Price'].str.replace('Rp', '', regex=False)
df['Price'] = df['Price'].str.replace('.', '', regex=False)
df['Price'] = df['Price'].str.strip()
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')


# Hapus data yang harga-nya kosong atau aneh
df = df.dropna(subset=['Price'])

# Reset index
df.reset_index(drop=True, inplace=True)

# Cek hasil preprosessing
print(df.info())
print(df.describe())
print(df.head)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Misalkan df adalah DataFrame yang sudah ada
# df = pd.read_csv('data_mobil.csv')  # Contoh cara memuat data

# Pisahkan fitur dan target
X = df.drop(columns=['Harga', 'Nama Mobil'])
y = df['Harga']

# Define categorical features for one-hot encoding
categorical_features = ['Merk', 'Bahan Bakar', 'Tranmisi']

# Create a ColumnTransformer to apply different preprocessing to different columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', ['Tahun', 'Kilometer']),  # Numerical features
        ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features)  # Categorical features
    ])

# Fit and transform the data
X_transformed = preprocessor.fit_transform(X)

# Get feature names after transformation
feature_names = preprocessor.get_feature_names_out()

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.2, random_state=42)

# Model awal untuk feature importance
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)

# Seleksi fitur otomatis
selector = SelectFromModel(rf, threshold='median')
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

# Mendapatkan importances
importances = pd.Series(rf.feature_importances_, index=feature_names)

# Plotting feature importance
importances.sort_values().plot(kind='barh', figsize=(10, 6), title="Feature Importance")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np



# Pisahkan fitur dan target
X = df.drop(columns=['Harga', 'Nama Mobil'])
y = df['Harga']

# Define categorical features for one-hot encoding
categorical_features = ['Merk', 'Bahan Bakar', 'Tranmisi']

# Buat ColumnTransformer untuk preprocessing fitur numerik dan kategori
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', ['Tahun', 'Kilometer']),
        ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features)
    ])

# Transform fitur
X_transformed = preprocessor.fit_transform(X)

# Ambil nama fitur setelah transformasi
feature_names = preprocessor.get_feature_names_out()

# Bagi data train dan test
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.2, random_state=42)

# Latih model Random Forest Regressor
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)

# Seleksi fitur otomatis berdasarkan feature importance median
selector = SelectFromModel(rf, threshold='median')
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

# Latih ulang model dengan fitur terpilih
rf_selected = RandomForestRegressor(random_state=42)
rf_selected.fit(X_train_selected, y_train)

# Prediksi harga mobil pada data test
y_pred = rf_selected.predict(X_test_selected)

# Evaluasi model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Evaluasi Model Random Forest Regressor (fitur terpilih):")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R-squared (R2 Score): {r2:.2f}")

# Visualisasi hasil prediksi vs aktual
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', linewidth=2)
plt.xlabel('Harga Aktual')
plt.ylabel('Harga Prediksi')
plt.title('Plot Harga Aktual vs Harga Prediksi Mobil Bekas')
plt.show()

```
